2. Create a plot of the function; label the axes and make a title. Note that depending on your student number, you may have one, two, three or four minima -- be sure to choose a domain for your plot so that all of the minima are shown.

```python
A = 0
B = 8
C = 9
D = 1

def myfunc(a,b,c,d):
    y = (x-a)*(x-b)*(x-c)*(x-d)
    return y
```

```python
# 100 linearly spaced numbers
x = np.linspace(-4,18,100)

# plot the function
plt.ylim(top=320)
plt.ylim(bottom=-30)
plt.plot(x,myfunc(A,B,C,D))

# show the plot
plt.title("Quartic Function")
plt.xlabel("Values of x")
plt.ylabel("Values of y")
plt.show()
```

3. Construct the loss function as a lambda function. The loss function will be the equation given above.

```python
loss=lambda:myfunc(A,B,C,D)
```

4. Now, use the Adam optimizer to find the minimum of the loss function.

```python
x = tf.Variable(0.0)
```

```python
optimizer=tf.optimizers.Adam(.01)
```

```python
iterations=[]
loss_values=[]
for i in range(200):
    optimizer.minimize(loss,x)
    iterations.append(i)
    loss_values.append(loss())

plt.plot(iterations, loss_values)
plt.title("loss vs iterations")
plt.xlabel("iterations")
plt.ylabel("loss")
plt.show()
```

```python
tf.print(x)
```

5. Use the SGD optimizer

```python
x = tf.Variable(0.0)
```

```python
optimizer_SGD=tf.optimizers.SGD(.01)
```
